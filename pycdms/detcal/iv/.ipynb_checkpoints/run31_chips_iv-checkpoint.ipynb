```python
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import numpy as np
import pickle
import os

from ivtools import IV
```

```python
Rshunt = 5.0e-3
dRshunt = 0.01 * Rshunt
Rbias_SG = 20000.0
Rfb = 5000.0
loopgain = 2.4
ADCperVolt = 65536.0/2.0
currentGUItoReal = 1.0
sampleRate = 625.0e3

# 800x200 F
# 200x50 A
# 400x100 B
channelList = ['PFS1','PAS2','PBS2']
DCRC = ['DCRC2','DCRC1','DCRC1']
Rp = np.array([7.1e-3,0.0112022046128-Rshunt,7.3e-3])
dRp = np.array([6.24781130475e-08,6.24781130475e-08,6.24781130475e-08])

saveProcessedPkl = False


SeriesFolders = ['19mK/',
               '24mK/',
               '29mK/',
               '34mk/',
               '38mK/',
               '45mK/',
               '51mK/',
               '57mK/']

SeriesID = ['19mK',
               '24mK',
               '29mK',
               '34mK',
               '38mK',
               '45mK',
               '51mK',
               '57mK']

nSeries = len(SeriesFolders)

pathSavePlots='/nervascratch/samwatkins/data/Run31/TES_chips/Noise/plots/'

pathToPkl = '/nervascratch/samwatkins/data/Run31/TES_chips/Noise/'
pathSaveProcessed = '/nervascratch/samwatkins/data/Run31/TES_chips/Noise/processed/'
    
    
nChan = 3
chanColor = ['b','g','r']

Vboffset=np.zeros(nSeries)
dVboffset=np.zeros(nSeries)

```

```python
idArray = [list() for __ in range(nSeries)]



############ get Rn, Offsets, Ptes, R0 for each channel
Is_uncal = list()
Is = list()
Ptes = list()
R0 = list()
Vbias = list()

dIs_uncal = list()
dR0 = list()
dIs = list()
dPtes = list()
dVbias = list()

Rn = list()
Rl = list()
dRn = list()
dRl = list()
Isoffset = list()
dIsoffset = list()
```

```python
for iSeries in  range(0,nSeries):
    pathToData = '/nervadata/SLAC/Run31/byseries/TES_chips/' + SeriesFolders[iSeries]
    
    pathToQETbias = '/nervadata/SLAC/Run31/byseries/TES_chips/' + SeriesFolders[iSeries]
        
    idArrayAll = os.listdir(pathToData)[:-1]
    nFiles = len(idArrayAll)
    
    for iFile in range(0,nFiles):
        if np.mod(iFile,4)==0:
            idArray[iSeries].append(idArrayAll[iFile])
    
    nIters = len(idArrayAll)//4
    
    # temp variables
    _Is_uncal = np.zeros((nChan,nIters))
    _Is = np.zeros_like(_Is_uncal)
    _Ptes = np.zeros_like(_Is_uncal)
    _R0 = np.zeros_like(_Is_uncal)
    _Vbias = np.zeros_like(_Is_uncal)

    _dIs_uncal = np.zeros_like(_Is_uncal)
    _dR0 = np.zeros_like(_Is_uncal)
    _dIs = np.zeros_like(_Is_uncal)
    _dPtes = np.zeros_like(_Is_uncal)

    _Rn = np.zeros(nChan)
    _Rl = np.zeros_like(_Rn)
    _dRn = np.zeros_like(_Rn)
    _dRl = np.zeros_like(_Rn)
    _Isoffset = np.zeros_like(_Rn)
    _dIsoffset = np.zeros_like(_Rn)

    qetBiasInfo = os.listdir(pathToData)[-1]
    qetBias = np.load(pathToQETbias + qetBiasInfo)[:,:nIters] * 1e-6
    
    _Vbias = qetBias*Rshunt + Vboffset[iSeries]
    _dVbias = np.sqrt(qetBias**2.0 * dRshunt**2.0 + dVboffset[iSeries]**2.0)

    # from dIdV fit
    _Rl = Rp + Rshunt
    _dRl = np.sqrt(dRp**2.0+dRshunt**2.0)
    
    for iIter in range(0,nIters):
            for iChan in range(0,nChan):
                
                fnamePkl = 'Noise'+DCRC[iChan] + '_'+ channelList[iChan] +'_' + idArray[iSeries][iIter]
                
                with open(pathToPkl + fnamePkl + '.pkl','rb') as f:
                    dataIV=pickle.load(f, encoding="bytes")
    
                _Is_uncal[iChan,iIter] = dataIV[b'Is0']
                _dIs_uncal[iChan,iIter] = dataIV[b'dIs0']

    
    Is_uncal.append(_Is_uncal)
    dIs_uncal.append(_dIs_uncal)
    
    Vbias.append(_Vbias)
    dVbias.append(_dVbias)

    Rl.append(_Rl)
    dRl.append(_dRl)
    
```

```python
maxd = np.max([Is_uncal[i].shape[-1] for i in range(len(Is_uncal))])
dIs = np.empty((len(Is_uncal),3,maxd))
dIs[:] = np.nan
dIs_err = np.copy(dIs)
Vb = np.copy(dIs)
Vb_err = np.copy(dIs)

for i in range(len(Is_uncal)):
    dIs[i,:,:Is_uncal[i].shape[-1]] = Is_uncal[i]
    dIs_err[i,:,:dIs_uncal[i].shape[-1]] = dIs_uncal[i]
    Vb[i,:,:Vbias[i].shape[-1]]= Vbias[i]
    Vb_err[i,:,:dVbias[i].shape[-1]] = dVbias[i]

Rload = Rl[0]
dRload = dRp[:]
    
normalinds = range(0,3)
```

```python
varsave = [dIs, dIs_err, Vb, Vb_err, Rload, dRload]
varnames = ["dites", "dites_err", "vb", "vb_err", "rload", "rload_err"]

np.savez("test_iv_data", dites=dIs, dites_err=dIs_err, vb=Vb, vb_err=Vb_err, rload=Rload, rload_err=dRload)
```

```python
ivdata = IV(dIs,dIs_err,Vb,Vb_err,Rload,dRload,normalinds=normalinds)
ivdata.calc_iv()
```
